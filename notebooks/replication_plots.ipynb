```python
x = [1,2,3,4,5,6,7,8]
x2 = [1,2,3,4,8]
y = [2777,3182,3225,3204,3219,3224,3231,3235]
y2 = [2738,3146,3180,3201,3235]
y3 = [2667,3136,3178,3169,3237]
plt.figure(figsize=(7,5))
plt.plot(x, y, linestyle='--', marker='o', color='b', label='Num replicas=1')
plt.plot(x2, y2, linestyle='--', marker='o', color='r', label='Num replicas=2')
plt.plot(x2, y3, linestyle='--', marker='o', color='g', label='Num replicas=3')
plt.xlabel('Number of clients')
plt.ylabel('Throughput (reqs/s)')
plt.title('Chat Application Benchmark (with Multi-threaded Replication)')
plt.legend()
plt.show()
```
