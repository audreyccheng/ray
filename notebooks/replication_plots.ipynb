```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 16})
```

```python
x = [0,1710,2871,11251] #,385331
times = [0.0019507408142089844*1000,0.0025768280029296875*1000,0.003906965255737305*1000,0.004673004150390625*1000] #,0.011281728744506836
plt.figure(figsize=(7,5))
plt.plot(x, times, linestyle='--', marker='o', color='green')
plt.xlabel('Size of kv-store')
plt.ylabel('Recovery time (ms)')
plt.title('Recovery Time From Backup')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [2777,3182,3225,3204,3219,3224,3231,3235]
plt.figure(figsize=(7,5))
plt.plot(x, y, linestyle='--', marker='o')
plt.xlabel('Number of clients')
plt.ylabel('Throughput (reqs/s)')
plt.title('Chat Application Benchmark (with Multi-threaded Replication)')
plt.show()
```

```python
x = [1,2,4]
y1 = [2404,2332,2405]
y2 = [2812,3236,3375]
y3 = [2146,2244,2257]
plt.figure(figsize=(7,5))
plt.plot(x, y1, linestyle='--', marker='o', color='b', label='Single-threaded')
plt.plot(x, y2, linestyle='--', marker='o', color='g', label='Multi-threaded (10 threads)')
plt.plot(x, y3, linestyle='--', marker='o', color='r', label='Async')
# locs, labels = plt.xticks()
plt.xticks(np.arange(1, 5, step=1))
plt.xlabel('Number of clients')
plt.ylabel('Throughput (reqs/s)')
plt.title('Different Replication Implementations')
plt.legend()
plt.show()
```
